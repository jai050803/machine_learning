## Linear Regression

Draw a straight line beween the data points using the relationship. This line can be used to predict the future values

#### How does it work

1. Draw a scatter plot between the data

2. Import scipy and draw the line of linear regression between them 

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))
plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()
```

```python

print(round(r,2))
print(round(p,2))
print(round(std_err,2))
```

The equation for the line is y*x + b where b is the intercept.

## R is for relationship

If there is no relationship then the linear regression model cannot predict anything 

the value of r ranges from -1 to 1 where 0 means no relationship and 1 and -1 means 100% related

In above data - the relationship r is 76%, not perfect but can be used for linear regression.

now we can use this data to predict the future values

```python
speed = myfunc(10)
print(round(speed,2)) # give the speed for a 10 year old car
```

## Bad Fit and Good Fit

bad fit is the data where the linear regression model is not the perfect fit to predict the future values

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
print(round(r,3))
```

as the relationship is just 0.013 which means a very bad relationship and hence we cannot predict the outcomes with the values and hence not suitable for the linear regression model

## Polynomial Regression

If your data points clearly will not fit a linear regression (a straight line through all data points), it might be ideal for polynomial regression.

Polynomial regression, like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points.
