## Machine Learning Day 1 - Introduction

#### what is machine learning?

Machine learning is programming computers to optimize a performance criterion using example 
data or past experience. We have a model defined up to some parameters, and learning is the 
execution of a computer program to optimize the parameters of the model using the training data or 
past experience. The model may be predictive to make predictions in the future, or descriptive to gain 
knowledge from data, or both.

## Data Types

1. Numerical data - data which are in form of numbers. it have two categories -  
        a. Discrete Data - Limited to integers  
        b. Continous Data - can be any number

2. Categorial Data - Values that cannot be measured up against each other

3. Ordinal data - just like categorial data but can be measured up against each other

## Types of Machine Learning

### Supervised Learning 

The goal is to map the input variable x with the output variable y

#### Steps of supervised learning

1. labelled dataset  
2. Model Training  
3. Predictions
4. Output


### Types of Supervised Machine learning algorithms

    1. Regression
    2. Classification

#### Regression 
In regression, We take the numeric data. It is used when the independent and the dependent variables have some relation between them.

Data taken is continous like data of weight where weights don't have much difference or the data of the market, etc.

##### Some common regression algorithms used are 

1. Linear Regression
2. Polynomial regression
3. Regression Trees

### Classification

It is used on the categorial data. Categorial data are those which have 2 logics like yes/no or true/false or right/wrong or day/night etc. These are called binary classification and if we have multiple output values then we call it multiple classification.

##### Some common classification techniques are 

1. Random Forest
2. Decision Trees
3. Logistic Regression
4. Support Vector machines

## Advantages of Supervised Machine Learning

1. You have full control over what the machine is learning
2. You can easily test and debug your model
3. You can determine the number of classes

## disadvantages of Supervised machine learning

1. Limited scope
2. Collecting labelled data is time consuming
3. Wrong prediction
