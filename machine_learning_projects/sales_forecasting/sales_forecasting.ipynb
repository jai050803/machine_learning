# Sales Forecasting

### Importing the libraries

```python
import pandas as pd 
import numpy as np 
```

### Loading the dataset

```python
filename = 'sales.csv'
data = pd.read_csv(filename)
data.head()
```

### Checking for the data Types of the columns

```python
data.info
```

### finding the mean of the data

```python
data.mean()
```

### Handling the missing values

```python

data['rate'].fillna(0, inplace=True)
data.head()
```

```python
data['sales_in_first_month'].fillna(118, inplace=True)
data.head()
```

### Converting the data type 

```python

data['sales_in_first_month'] = data['sales_in_first_month'].astype('int64')
data['rate'] = data['rate'].astype('int64')
data.info()
```

```python
data.head()
```

### Feature Selection

```python
X = data.iloc[:, :3]
Y = data.iloc[:, -1]
```

```python
X.head()
```

```python
Y.head()
```

### Concatenating X and Y

```python
df = pd.concat([X, Y], axis=1)
df.head()
```
