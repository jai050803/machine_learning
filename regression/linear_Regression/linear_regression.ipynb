```python
# Importing the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

as the data do not have any empty values or missing values.. so we will continue with creating the testing and training set for the data

```python
X = df.iloc[:, :-1].values  #creating array of independent variables
y = df.iloc[:, 1].values  #creating array of dependent variables
```

```python
# creating the testing and the training set
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```
