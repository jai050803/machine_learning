# Simple Linear Regression

Step 1 : Apply the data preprocessing over the data

```python
# Importing the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

as the data do not have any empty values or missing values.. so we will continue with creating the testing and training set for the data

```python
X = df.iloc[:, :-1].values  #creating array of independent variables
y = df.iloc[:, 1].values  #creating array of dependent variables
```

```python
# creating the testing and the training set
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

#### Fitting the Simple linear Regression algorithm to the training set

```python
# Importing the library

from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression() #creating the object regressor
regressor.fit(X_train, Y_train) #fitting the object to the training set
```

##### Creating the y_pred vector which contains predicted salary for all the observations of test set

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
#comparing the y_pred and the y_test data to see how much the prediction is accurate

print(Y_test)
print(y_pred) 
```

#### Visualizing the training set results

```python
plt.scatter(X_train, Y_train, color = 'red', marker = '*')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Simple linear regression Model")
plt.xlabel('Years of Experience')
plt.ylabel('salary')
plt.show()
```

```python
# visualizing the test set results
plt.scatter(X_test, Y_test, color = 'red', marker = '*')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Simple linear regression Model")
plt.xlabel('Years of Experience')
plt.ylabel('salary')
plt.show()
```
