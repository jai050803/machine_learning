# Polynomial Regression

```python
#importing the libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('Position_Salaries.csv')

X = df.iloc[:, 1:2].values
y = df.iloc[:, -1].values
```

1. we don't need to split the data into training and the testing set.
we also need not to apply the feature scaling to the data

#### Fitting the linear regression to the dataset

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
plt.scatter(X, y , color="hotpink", marker= "+")
plt.plot(X, regressor.predict(X), color="blue")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

#### Fitting the polynomial regression to the dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
X_poly
```

```python
# fit this X_poly with linear regression

lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
#visualize the result of the polynomial regression
plt.scatter(X, y , color="hotpink", marker= "+")
plt.plot(X, lin_reg2.predict(X_poly), color="blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```
